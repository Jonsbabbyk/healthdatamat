```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values between 0 and 2*pi, step size of 0.5
x = np.arange(0, 2 * np.pi, 0.5)

# Compute the sine of each x value
y = np.sin(x)

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b')  # Adding markers and line style

# Add title and labels
plt.title("Sine Wave with Specific Values")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")

# Display the grid
plt.grid(True)

# Show the plot
plt.show()

# Print x and y values
print("x values:", x)
print("y values (sin(x)):", y)
```

```python
import matplotlib.pyplot as plt

# Total population
total_population = 1000000

# Categories and their respective numbers based on total population
categories = [
    "Critical Diseases (50%)",  # 500,000 people
    "Common Diseases (10%)",  # 100,000 people
    "Accidents (15%)",  # 150,000 people
    "Maternity (5%)",  # 50,000 people
    "Random Checkups (15%)",  # 150,000 people
    "Unknown (5%)"  # 50,000 people
]

# Healthcare visits before the funding change (initial data for 1 million people)
before_funding = [500000, 100000, 150000, 50000, 150000, 50000]

# Healthcare visits after funding change (only 10% can access electronic payments)
after_funding = [0.1 * x for x in before_funding]  # Only 10% of the population can access electronic payments

# Critical Impact (The reduced number of healthcare visits after funding change)
critical_impact = [x - y for x, y in zip(before_funding, after_funding)]  # The difference showing the critical drop

# Create a figure with 3 subplots
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# Graph 1: Before Funding Change
ax[0].bar(categories, before_funding, color='royalblue')
ax[0].set_title('Before Funding Change (Cash Payment)')
ax[0].set_xlabel('Health Categories')
ax[0].set_ylabel('Number of People')
ax[0].set_ylim(0, total_population)

# Graph 2: After Funding Change
ax[1].bar(categories, after_funding, color='orange')
ax[1].set_title('After Funding Change (10% Access to Electronic Payments)')
ax[1].set_xlabel('Health Categories')
ax[1].set_ylabel('Number of People')
ax[1].set_ylim(0, total_population)

# Graph 3: Critical Impact (Difference between Before and After Funding Change)
ax[2].bar(categories, critical_impact, color='darkred')
ax[2].set_title('Critical Impact: Reduction in Healthcare Access Due to Payment Change')
ax[2].set_xlabel('Health Categories')
ax[2].set_ylabel('Difference in Number of People')
ax[2].set_ylim(0, max(critical_impact) + 50000)

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```
